```python
import cv2 
capture = cv2.VideoCapture(0)
fourcc = cv2.VideoWriter_fourcc(*'XVID')
output = cv2.VideoWriter('Movie.avi',fourcc,20.0 , (640,480))
print(capture.isOpened())
while(capture.isOpened()):
    ret , frame = capture.read()
    if ret == True:
        print(capture.get(cv2.CAP_PROP_FRAME_WIDTH))
        print(capture.get(cv2.CAP_PROP_FRAME_HEIGHT))
        output.write(frame)
        gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        cv2.imshow('frame',gray)
        if cv2.waitKey(1) & 0xff==ord('q'): break 
    else : break 
capture.release()
output.release()
cv2.destroyAllWindows()
```

```python
import cv2

# فتح الكاميرا
cap = cv2.VideoCapture(0)

# التحقق من استعداد الكاميرا
if not cap.isOpened():
    print("لا يمكن فتح الكاميرا")
    exit()

while True:
    # قراءة الإطار الحالي من الكاميرا
    ret, frame = cap.read()

    # التحقق من نجاح قراءة الإطار
    if not ret:
        print("لا يمكن قراءة الإطار من الكاميرا")
        break

    # عرض الإطار المقروء
    cv2.imshow('Camera', frame)

    # الانتظار لضغط مفتاح 'q' لإنهاء البرنامج
    if cv2.waitKey(1) == ord('q'):
        break

# إغلاق الكاميرا وإغلاق النوافذ
cap.release()
cv2.destroyAllWindows()
```
